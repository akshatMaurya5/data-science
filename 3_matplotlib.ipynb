```python
import matplotlib.pyplot as plt
%matplotlib inline  # For Jupyter Notebooks (displays plots inline)
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [15, 20, 10, 25]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

plt.hist(data, bins=5, color='green', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

customizing plots

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.legend()
plt.grid(True)
plt.show()
```

subplots


```python
plt.figure(figsize=(10, 5))  # Width, Height

# Subplot 1
plt.subplot(1, 2, 1)  # (rows, columns, index)
plt.plot(x, y1, color='red')
plt.title("Plot 1")

# Subplot 2
plt.subplot(1, 2, 2)
plt.scatter(x, y2, color='blue')
plt.title("Plot 2")

plt.tight_layout()  # Prevents overlapping
plt.show()
```
